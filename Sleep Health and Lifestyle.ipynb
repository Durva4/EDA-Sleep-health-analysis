Q1. What is the distribution of Gender in the dataset?

```python
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Percentage Distribution")
plt.show()
```

##### Inference:
The dataset exhibits a near-equal distribution of gender, with Males comprising 50.5% and Females comprising 49.5%.

Q2.What is the distribution of Age?


```python
sns.histplot(df['Age'])
plt.title("Age Distribution")
plt.show()
```

##### Inference:
The age distribution is multi-modal and not uniform modal, with the highest concentration of participants falling around the 45-50 age group.

Q3. What is the minimum and maximum number of Daily Steps taken by males compared to females?

```python
s = df.groupby('Gender')['Daily Steps'].agg(['min', 'max']).reset_index()
s.columns = ['Gender', 'Minimum Steps', 'Maximum Steps']
print(s)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Daily Steps', data=df, palette='Set2')
plt.title('Range of Daily Steps: Males vs Females', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Daily Steps', fontsize=12)
plt.show()
```

##### Inference:
Males generally take slightly more daily steps with greater variability, while females show more consistent step counts with occasional high-activity outliers.

Q4. What is the average level of stress by gender?

```python
sns.barplot(x='Gender', y='Stress Level', data=df,errorbar=None)
plt.title("Stress Level by Gender")
plt.show()
```

##### Inference:
The average stress level for males is visibly higher than the average stress level for females.

Q5. What is daily Steps count across BMI categories?

```python
sns.barplot(x='BMI Category', y='Daily Steps', data=df,errorbar=None)
plt.title("Daily Steps by BMI Category")
plt.show()
```

##### Inference:
The Obese category has a significantly lower average Daily Steps count compared to the Overweight, Normal, and Normal Weight categories.


Q6. How is Sleep Duration influenced by BMI Category?

```python
df['BMI Category'].unique()
```

```python
sns.boxplot(x='BMI Category', y='Sleep Duration', data=df)
plt.title("Sleep Duration by BMI Category")
plt.show()
```

Q7. How do Stress Levels vary across different Occupation groups?

```python
sns.barplot(x='Occupation', y='Stress Level', data=df,errorbar=None)
plt.xticks(rotation=85)
plt.title("Stress Level by Occupation")
plt.show()
```

##### Inference:
Sales Representatives and Scientists report the highest average stress levels, while Engineers and Accountants report the lowest average stress levels.


Q8. How does Gender vary across different Occupations?

```python
pd.crosstab(df['Occupation'], df['Gender']).plot(kind='bar')
plt.title("Occupation vs Gender")
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

##### Inference:
Nursing is overwhelmingly female-dominated, while Doctor, Lawyer, Salesperson, and Software Engineer occupations are heavily dominated by males in this dataset.

Q9. How many people fall into each sleep quality level within overweight category?

```python
df['Quality of Sleep'].unique()
```

```python
overweight_df = df[df['BMI Category'] == 'Overweight']
sleep_counts = overweight_df['Quality of Sleep'].value_counts().sort_index()
print("Counts of Quality of Sleep for Overweight category:")
print(sleep_counts)
plt.figure(figsize=(10, 6))
sns.countplot(data=overweight_df, x='Quality of Sleep', hue='Quality of Sleep', palette='magma', legend=False)

plt.title("Distribution of Quality of Sleep (Overweight Category)")
plt.xlabel("Quality of Sleep Score")
plt.ylabel("Number of People")
plt.show()
```

##### Inference:
The majority of individuals in the overweight category experience mediocre sleep quality, with a primary peak at a score of 6.

Q.10 How does 'Physical Activity Level' vary across different 'Age' groups?

```python
age_activity = df.groupby(pd.cut(df['Age'], bins=5))['Physical Activity Level'].mean().round(2)
plt.figure(figsize=(10, 6))
age_activity.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("Mean Physical Activity Level by Age Group", fontsize=14)
plt.xlabel("Age Range", fontsize=12)
plt.ylabel("Mean Physical Activity Level", fontsize=12)
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

#### Inference:
The highest mean Physical Activity Level is observed in the age 46 to 52, suggesting activity peaks in middle age before slightly declining.

Q11. Analyze how Blood Pressure varies with Sleep Duration and Physical Activity Level.

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Sleep Duration'], df['Blood Pressure'], 
            c=df['Physical Activity Level'], cmap='viridis', s=80)

plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Blood Pressure")
plt.title("Blood Pressure vs Sleep Duration")
plt.colorbar(label='Physical Activity Level')
plt.show()
```

##### Inference:
There is no clear linear relationship between Sleep Duration and Blood Pressure; however, the highest Physical Activity Levels (bright yellow/green points) tend to cluster around higher Sleep Durations.

Q12.What is the mean 'Age' for the top 3 'Occupation' types by count?

```python
a= df['Occupation'].value_counts().head(3).index
df[df['Occupation'].isin(a)].groupby('Occupation')['Age'].mean().round(1)
```

##### Inference:
The top three most frequent occupations show a clear age progression: Doctors are the youngest, followed by Engineers, and then Nurses are the oldest.
